```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px 
pd.set_option('display.max_columns', 1000)

#import des diffèrentes librairies ainsi que des options de lecture
```

```python
df = pd.read_csv("src/Speed Dating Data.csv", encoding='ISO-8859-1')
#lecture du fichier que l'on appelle df
```

Première analyse du fichier

```python
# Basic stats
print("Number of rows : {}".format(df.shape[0]))
print()

print("Display of dataset: ")
display(df.head())
print()

print("Basics statistics: ")
data_desc = df.describe(include='all')
display(data_desc)
print()

print("Percentage of missing values: ")
display(100*df.isnull().sum()/df.shape[0])
```

Première constatation:
Beaucoup de colonnes avec un fort pourcentage de valeurs manquantes (inexploitables pour certaines)

Données écartées

```python
# je décide d'écarter les waves 6 à 9 car les données saisies lié à l'expérience mettent un biais dans les résultats 
# à savoir les autres waves sont évaluées sur un répartition de 100 point à répartir (et donc faire des choix ayant des 
#incidences sur les autres variables), tandis que sur les wave 6 à 9, la répartition des points se fait par variable et
#n'influe pas les variables entre elles.
mask = (df["wave"] < 10) & (df["wave"] > 5)
df = df[~mask]
```

```python
df.describe(include='all')
```

```python
#transformation des données, on transforme la colonne genre en 0 et 1 en Woman et Man pour une meilleur visualisation dans la suite des graphiques
df['gender'] = df['gender'].apply(lambda x : 'Woman' if x == 0 else 'Man')
df['goal'].replace(1.0, "Seemed like a fun night out", inplace = True)
df['goal'].replace(2.0, "To meet new people", inplace = True )
df['goal'].replace(3.0, "To get a date", inplace = True )
df['goal'].replace(4.0, "Looking for a serious relationship", inplace = True )
df['goal'].replace(5.0, "To say I did it", inplace = True )
df['goal'].replace(6.0, "Other", inplace = True )
```

Création d'une seconde dataframe pour extraires les données par personnes

```python
#pour la représentation graphique on choisi de créer une nouvelle df avec les personnes uniques:
you_look_opposite_sex = ['attr1_1', 'sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']
you_think_other_want = ['attr2_1', 'sinc2_1', 'intel2_1', 'fun2_1', 'amb2_1', 'shar1_2']
your_quality = ['attr3_1', 'sinc3_1', 'intel3_1', 'fun3_1', 'amb3_1']
hobbies = ['sports', 'tvsports', 'exercise', 'dining', 'museums', 'art', 'hiking', 'gaming', 'clubbing', 'reading', 'tv', 'theater', 'movies', 'concerts', 'music', 'shopping', 'yoga']
df_unique = df[['iid', 'wave'] + ['age'] + ['goal'] + ['gender'] + ['exphappy'] + you_look_opposite_sex + you_think_other_want + your_quality +hobbies ].drop_duplicates()
df_unique.head()
```

Présentation des personnes

```python
# graphique superposé pour représenter la répartition de l'age par rapport au genre
fig = sns.FacetGrid(df_unique, hue='gender', height= 5, palette= ['pink','blue'])
fig.map(plt.hist, 'age', bins=10)
fig.set_xlabels('Age')
plt.subplots_adjust(top=0.9)
fig.fig.suptitle("Age distribution by gender")
fig.add_legend()
plt.show()
```

```python
#visualisation des objectifs en début de soirée pour les hommes, focus sur ceux qui cherchent un rdv ou un relation sérieuse
labels = ["Seemed like a fun night out", "To meet new people", "To get a date", "Looking for a serious relationship",
          "To say I did it", "Other"]

fig = go.Figure(data=[go.Pie(labels=labels, values=df_unique[df_unique['gender'] == 'Man']["goal"].value_counts(), textinfo='label+percent',
                             insidetextorientation='radial',
                             pull=[0, 0, 0.2, 0.2, 0., 0]
                            )])
fig.update(layout_title_text='Primary goal in participating in this event for Men',
          layout_legend=dict(yanchor="top", y=0.6, xanchor="left", x=-1
))

fig.show()
```

```python
#visualisation des objectifs en début de soirée pour les femmes, focus sur celles qui cherchent un rdv ou un relation sérieuse
labels = ["Seemed like a fun night out", "To meet new people", "To get a date", "Looking for a serious relationship",
          "To say I did it", "Other"]

fig = go.Figure(data=[go.Pie(labels=labels, values=df_unique[df_unique['gender'] == 'Woman']["goal"].value_counts(), textinfo='label+percent',
                             insidetextorientation='radial',
                             pull=[0, 0, 0.2, 0.2, 0., 0]
                            )])
fig.update(layout_title_text='Primary goal in participating in this event for Women',
          layout_legend=dict(yanchor="top", y=0.6, xanchor="left", x=-1
))

fig.show()
```

Résultat du questionnaire avant la soirée

```python
#attente chez le sexe opposé
df_unique[['gender'] + you_look_opposite_sex].groupby(['gender']).mean().plot(kind='bar')
plt.title("what you look for in the opposite sex")
```

```python
# ce que vous pensez que le sexe opposé attend
df_unique[['gender'] + you_think_other_want].groupby(['gender']).mean().plot(kind='bar')
plt.title("What do you think the opposite sex looks for in a date")
```

```python
#vos qualités
fig = df_unique[['gender'] + your_quality].groupby(['gender']).mean().plot(kind='bar')
plt.title("How do you think you measure up")
```

Passions des personnes: un sujet de discussion pour le premier rendez-vous?

```python
#matrice de corrélation sur les passions
corr = df_unique[hobbies].corr()
f, ax = plt.subplots(figsize=(20, 8))
fig = sns.heatmap(corr, annot=True, ax=ax, cmap= 'seismic')
plt.title('Correlation with hobbies', fontsize=25)
plt.show()
```

Analyse par vague des décisions par genre

```python

fig = px.bar(data_frame= df, x= 'order',y='dec', color='gender' , barmode="group", animation_frame='wave', range_x=[0,23], range_y=[0,30], 
category_orders={'gender': ['Man', 'Woman']}, title='Sum of positive decison by dating in each wave by gender')
fig.show()
```

Correlation entre les résultats des questionnaires avant la soirée et la possibilité d'un second rendez-vous

```python
#corrélation entre les notations d'intérêts du participant et du partenaire avant la soirée et le match
fig = px.scatter(df, x="order", y="int_corr",color='match', animation_frame='wave',range_x=[0,23], range_y=[-1,1],
 title="Correlation between participant's and partner's ratings of interests before the speed dating night and match decision")
fig.show()
```
